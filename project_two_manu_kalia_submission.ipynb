````python
# Root Mean Square Error Function
def RMSE(true, predicted):
    squared_diff = np.square(true - predicted)    
    return np.sqrt(np.mean(squared_diff))
````

### First scenario... set up X and y, cross-train-split

````python
X1 = df[features]
y1 = df['SalePrice']
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=22)

X1_kag = testdf[features]
````

### Set up first Pipeline, `pipe1_lasso`, to instantiate, fit, and score a scaled Lasso model

````python
steps1_lasso = [
     ('scaler', StandardScaler()),
     ('lasso', LassoCV(cv=5))
    ]

pipe1_lasso  = Pipeline(steps = steps1_lasso)
pipe1_lasso.fit(X1_train, y1_train)

pred1_lasso = pipe1_lasso.predict(X1_test)

print(f'\n"pipe1_lasso" Train score: {pipe1_lasso.score(X1_train, y1_train):.6f}')
print(f'"pipe1_lasso" Test score: {pipe1_lasso.score(X1_test, y1_test):.6f}')
print(f'"pipe1_lasso" RMSE score: {RMSE(y1_test, pred1_lasso):.2f}\n')
````

````python
pred1_lasso[:25]
````

````python
#  Get the lasso coefficients
pipe1_lasso_regularizer = pipe1_lasso.named_steps['lasso']
pipe1_lasso_coeff = pd.DataFrame(pipe1_lasso_regularizer.coef_, columns=['lasso_coeff'],
                                 index=features).sort_values(by='lasso_coeff', ascending=True)
pipe1_lasso_coeff.shape
````

````python
plt.figure(figsize=(15,15))
plt.barh(pipe1_lasso_coeff.T.columns, pipe1_lasso_coeff['lasso_coeff'], color='goldenrod')
plt.xticks(rotation=90)
plt.title('All Lasso Coefficients - Pipe1 (73 Features)', fontsize=20)
plt.xlabel('Lasso Coefficient Values', fontsize=14)
plt.ylabel('Feature-set 1', fontsize=20)
# plt.savefig('./datasets/Pipe1_lasso_coeff.png')
plt.show();
````

### Implement `pipe1_lasso` on the Kaggle validation data, `testdf`, to generate `SalePrice` predictions for submission to Kaggle website

````python
pred1_lasso_kag = pipe1_lasso.predict(X1_kag)
pred1_lasso_kag_df = pd.DataFrame(pred1_lasso_kag, columns=['SalePrice'], index=testdf['Id'])

# pred1_lasso_kag_df.to_csv('./datasets/pred1b_lasso_kag.csv')
````
